```python
# Example new features
data['income_per_hour'] = data['capital_gain'] / (data['hours_per_week'] + 1)  # Prevent division by zero
data['capital_balance'] = data['capital_gain'] - data['capital_loss']
```

```python
# Strip whitespace from all columns
data = data.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Replace '?' with NaN
data.replace('?', pd.NA, inplace=True)
```

```python
data.isnull().sum()
```

```python
# Drop rows with missing values
cleaned_data = data.dropna()

# Check the new shape of the dataset
print(cleaned_data.shape)  # Should be smaller than (32561, 14)
```

```python
df=cleaned_data.copy()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Assuming 'cleaned_data' is your DataFrame

# List of columns already scaled and in data_standard_scaled
scaled_columns = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week','income_per_hour','capital_balance']

# Standard Scaling
scaler = StandardScaler()
data_standard_scaled = pd.DataFrame(scaler.fit_transform(df[scaled_columns]), columns=scaled_columns)

# Drop the scaled columns from cleaned_data
df = df.drop(columns=scaled_columns)


# Reset indices before concatenation 
df.reset_index(drop=True, inplace=True) 
data_standard_scaled.reset_index(drop=True, inplace=True)

# Concatenate the dataframes
final_data = pd.concat([df, data_standard_scaled], axis=1)


```

```python
df_min_max=data.copy()
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Assuming 'cleaned_data' is your DataFrame

# List of columns already scaled and in data_standard_scaled
scaled_columns_minmax = ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week','income_per_hour','capital_balance']

# Standard Scaling
minmax = MinMaxScaler()
data_scaled = pd.DataFrame(minmax.fit_transform(df_min_max[scaled_columns]), columns=scaled_columns_minmax)

# Drop the scaled columns from cleaned_data
df_min_max = df_min_max.drop(columns=scaled_columns_minmax)


# Reset indices before concatenation 
df_min_max.reset_index(drop=True, inplace=True) 
data_scaled.reset_index(drop=True, inplace=True)

# Concatenate the dataframes
final_data_minmax = pd.concat([df_min_max, data_scaled], axis=1)


```

When to use each scaling technique:

Standard Scaling: Preferred when the data follows a normal distribution. It centers the data around the mean with a unit standard deviation.

Min-Max Scaling: Preferred when the data does not follow a normal distribution. It scales the data to a fixed range (typically 0 to 1), preserving the relationship of data points.

```python
df2=final_data.copy()
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder


# Apply One-Hot Encoding to categorical variables with less than 5 categories
one_hot_cols = [col for col in df2.columns if df2[col].nunique() < 5]
one_hot_encoded_data = pd.get_dummies(df2, columns=one_hot_cols)

# Apply Label Encoding to categorical variables with more than 5 categories
label_encoder = LabelEncoder()
for col in df2.columns:
    if df2[col].dtype == 'object' and df2[col].nunique() >= 5:
        df2[col] = label_encoder.fit_transform(df2[col])

# Extract Label Encoded columns
label_encoded_cols = df2.columns[df2.dtypes == 'int32']

# Combine both one-hot and label encoded data
new_data = pd.concat([one_hot_encoded_data, df2[label_encoded_cols]], axis=1)


```

Pros and Cons of One-Hot and Label Encoding:

One-Hot Encoding:

Pros: No ordinal relationship assumption, suitable for nominal categorical variables.

Cons: Increases dimensionality, especially with high cardinality features.

Label Encoding:

Pros: Simple, does not increase dimensionality.

Cons: Implies ordinal relationships, which might not be accurate for nominal categories.

Discussion: PPS provides insights into feature relationships beyond linear correlations, detecting non-linear and asymmetric relationships, whereas correlation measures linear relationships.

```python
import ppscore as pps

# Calculate PPS matrix
pps_matrix = pps.matrix(df4)

# Calculate correlation matrix
correlation_matrix = df4.corr()

print(pps_matrix)

```

Insights from ppscore:
Predictive Relationships:

ppscore = 1.0: This typically indicates a perfect predictive relationship, meaning one variable predicts the other with no error. For example, when x = "age" and y = "age", the ppscore is 1.0, which is expected because a variable can perfectly predict itself.

ppscore = 0.0: This suggests no predictive relationship, meaning changes in one variable do not predict or correlate with changes in the other. For instance, age and fnlwgt (final weight) have a ppscore of 0.0, indicating no predictive relationship between them.
Pairings with High ppscore:

When x and y are the same variable (e.g., age predicting age), the ppscore will be 1.0. This is expected because a variable can always perfectly predict itself.
Similarly, in some cases where a variable has a strong, direct relationship with another, the ppscore could be close to 1. For example, if education_num directly correlates with the level of education, the ppscore might be high (closer to 1).
Model Metrics:

For some models or features, like the DecisionTreeRegressor, you may find that the ppscore for certain pairs is low (e.g., 0.0), meaning these features do not contribute much to the model's prediction ability. For example, variables like capital_gain and capital_loss may have low predictive power on education_num, and hence ppscore for such pairs might be close to 0.
Use in Feature Selection:

Variables with low ppscore (close to 0) are less useful for predicting other variables and may not be essential for model training.
Variables with higher ppscore values could be more useful for predictive tasks, as they show stronger relationships with other variables.
