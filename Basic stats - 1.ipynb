```python

df = numerical_cols['Volume']

plt.figure(figsize=(4, 5))
sns.boxplot(y=df, color='blue')  
plt.axhline(df.mean(), color='red', linestyle='--', label=f'Mean: {df.mean():.2f}')
plt.title(f'Box Plot of {df.name}')
plt.ylabel(df.name)
plt.legend()
plt.show()

# Calculate outliers
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df[(df < lower_bound) | (df > upper_bound)]

print("Lower Bound :",lower_bound,"Upper Bound :",upper_bound)
print(f"Outliers:\n{outliers}")
```

```python
df=pd.read_csv('sales_data_with_discounts.csv')
categorical_columns = df.select_dtypes(include=['object'])
categorical_columns.head(5)
```

```python
df=categorical_columns['BU'] #Business Unit COunt
plt.figure(figsize=(3, 3)) 
df.value_counts().plot(kind='bar') 
plt.title(f'Bar Chart for {df.name}') 
plt.xlabel(df.name) 
plt.ylabel('Count')
```

```python
df=categorical_columns['Brand'] #Brand count
plt.figure(figsize=(5, 3)) 
df.value_counts().plot(kind='bar') 
plt.title(f'Bar Chart for {df.name}') 
plt.xlabel(df.name) 
plt.ylabel('Count')
```

```python
df=categorical_columns['Model'] #Model Count
plt.figure(figsize=(10, 3)) 
df.value_counts().plot(kind='bar') 
plt.title(f'Bar Chart for {df.name}') 
plt.xlabel(df.name) 
plt.ylabel('Count')
```

```python

import matplotlib.pyplot as plt

grouped_data = df.groupby('BU')['Total Sales Value'].sum()

plt.figure(figsize=(5, 3)) 
grouped_data.plot(kind='bar', color=['blue', 'green', 'orange'])


plt.title('Total Sales by Business Unit')
plt.xlabel('Business Unit')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

```python

import matplotlib.pyplot as plt

grouped_data = df.groupby('BU')['Volume'].sum()

plt.figure(figsize=(5, 3)) 
grouped_data.plot(kind='bar', color=['blue', 'green', 'orange'])


plt.title('Total Volume by Business Unit')
plt.xlabel('Business Unit')
plt.ylabel('Total Volume')
plt.xticks(rotation=0)  
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


grouped_data = df.groupby('Model')['Total Sales Value'].sum()
plt.figure(figsize=(8, 6)) 
grouped_data.sort_values().plot(kind='barh', color=['blue', 'green', 'orange'])

plt.title('Total Sales by Model')
plt.xlabel('Total Sales Value')
plt.ylabel('Model')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


grouped_data = df.groupby('Model')['Volume'].sum()
plt.figure(figsize=(8, 6)) 
grouped_data.sort_values().plot(kind='barh', color=['blue', 'green', 'orange'])

plt.title('Volumes by Model')
plt.xlabel('Total Volume')
plt.ylabel('Model')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt


grouped_data = df.groupby('Brand')['Total Sales Value'].sum()
plt.figure(figsize=(7, 5)) 
grouped_data.sort_values().plot(kind='barh', color=['blue', 'green', 'orange'])

plt.title('Total Sales by Brand')
plt.xlabel('Total Sales Value')
plt.ylabel('Model')
plt.tight_layout()
plt.show()
```
